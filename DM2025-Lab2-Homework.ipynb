# **Competition Code**

## 1. Preprocessing Steps

```python
import pandas as pd
import json
import re
import numpy as np
import os

# 1. 讀取並解析 JSON (文字內容)
json_path = "./data/final_posts.json"
try:
    with open(json_path, 'r', encoding='utf-8') as f:
        raw_posts = json.load(f)
    
    parsed_posts = []
    for entry in raw_posts:
        try:
            post_data = entry['root']['_source']['post']
            parsed_posts.append({
                'id': post_data['post_id'],
                'text': post_data['text']
            })
        except KeyError:
            continue
            
    posts_df = pd.DataFrame(parsed_posts)
    print(f"Loaded {len(posts_df)} posts.")
except FileNotFoundError:
    print("Error: final_posts.json not found.")
    posts_df = pd.DataFrame(columns=['id', 'text'])

# 2. 讀取 CSV (ID 與 標籤)
try:
    data_id_df = pd.read_csv("./data/data_identification.csv")
    emotion_df = pd.read_csv("./data/emotion.csv")
except FileNotFoundError:
    print("Error: CSV files not found.")
    data_id_df = pd.DataFrame()
    emotion_df = pd.DataFrame()

# 3. 合併資料
# 先合併 split 資訊
merged = posts_df.merge(data_id_df, on="id", how="left")

# 分割 Train / Test
train_df = merged[merged["split"] == "train"].merge(emotion_df, on="id", how="left")
test_df  = merged[merged["split"] == "test"]

# 確保索引重置
train_df = train_df.reset_index(drop=True)
test_df  = test_df.reset_index(drop=True)

# 4. 文字清理
def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    # 轉小寫
    text = text.lower()
    # 移除 URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # 移除 @user
    text = re.sub(r'@\w+', '', text)
    # 移除多餘空白
    text = text.strip()
    return text

print("Preprocessing text...")
train_df['clean_text'] = train_df['text'].apply(preprocess_text)
test_df['clean_text'] = test_df['text'].apply(preprocess_text)

print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section
import torch
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer
import os

# 1. 設定模型名稱
model_name = "microsoft/deberta-v3-base"

print(f"Loading tokenizer for {model_name}...")
tokenizer = AutoTokenizer.from_pretrained(model_name)

# 2. 處理標籤 (Label Encoding)
labels_list = sorted(train_df["emotion"].unique())
label2id = {label: i for i, label in enumerate(labels_list)}
id2label = {i: label for label, i in label2id.items()}

# 將文字標籤轉為數字 ID
train_df["label"] = train_df["emotion"].map(label2id).astype(int)

# 3. 準備輸入資料
# 切分 10% 作為驗證集
X_train = train_df["clean_text"].tolist()
y_train = train_df["label"].tolist()
X_test_final = test_df["clean_text"].tolist()

train_texts, val_texts, train_labels, val_labels = train_test_split(
    X_train, y_train, test_size=0.1, random_state=42, stratify=y_train
)

# 4. Tokenization (文字轉向量)
max_len = 128
print("Tokenizing data...")

train_enc = tokenizer(train_texts, truncation=True, padding=True, max_length=max_len)
val_enc = tokenizer(val_texts, truncation=True, padding=True, max_length=max_len)
test_enc = tokenizer(X_test_final, truncation=True, padding=True, max_length=max_len)

# 5. 定義 Dataset 類別
class EmotionDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels=None):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        if self.labels is not None:
            item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.encodings['input_ids'])

# 建立 Dataset 物件
train_ds = EmotionDataset(train_enc, train_labels)
val_ds = EmotionDataset(val_enc, val_labels)
test_ds = EmotionDataset(test_enc)

print("Feature Engineering done.")
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
import torch

# 1. 定義評估函數 (計算準確率與 F1 Score)
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    f1 = f1_score(labels, preds, average='macro')
    acc = accuracy_score(labels, preds)
    return {'accuracy': acc, 'macro_f1': f1}

# 2. 初始化模型 model_name, labels_list, id2label, label2id
print(f"Initializing model: {model_name}")
model = AutoModelForSequenceClassification.from_pretrained(
    model_name, 
    num_labels=len(labels_list),
    id2label=id2label,
    label2id=label2id
)

# 3. 設定訓練參數
training_args = TrainingArguments(
    output_dir="./results_deberta",
    eval_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=2,
    num_train_epochs=3,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="macro_f1",
    fp16=torch.cuda.is_available(),
    report_to="none"
)

# 4. 初始化 Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

# 5. 開始訓練
print("Starting training...")
trainer.train()

# 6. 預測測試集
print("Predicting on test set...")
preds = trainer.predict(test_ds)
pred_labels = np.argmax(preds.predictions, axis=-1)

# 7. 轉換為文字標籤並儲存
pred_emotions = [id2label[i] for i in pred_labels]

submission = pd.DataFrame({
    'id': test_df['id'],
    'emotion': pred_emotions
})

submission_filename = 'submission_deberta_v3.csv'
submission.to_csv(submission_filename, index=False)
print(f"✅ {submission_filename} saved successfully! Please submit this file to Kaggle.")

# 顯示前幾筆預測結果確認格式
display(submission.head())
```
